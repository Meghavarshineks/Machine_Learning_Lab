```python
import numpy as np 
import pandas as pd
import seaborn as s
```

```python
data = pd.read_csv("/content/gdrive/MyDrive/Dataset/weatherHistory.csv")
data.head()
```

```python
data.shape
```

Preparing Data for training the

```python
data['Loud Cover'].value_counts()
```

```python
data.drop(['Formatted Date','Daily Summary','Loud Cover','Wind Bearing (degrees)'],axis=1,inplace=True)
```

```python
data
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['Precip Type'].value_counts()
```

```python
data
```

```python
data['Precip Type'].fillna(method='ffill',inplace=True,axis=0)
```

```python
data['Precip Type'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Precip Type']=le.fit_transform(data['Precip Type'])
data['Summary']=le.fit_transform(data['Summary'])
```

```python
data
```

```python
y =data.iloc[:,0]
```

```python
x = data.iloc[:,1:]
```

```python
x.corr() 
```

```python
x.drop('Apparent Temperature (C)',axis=1,inplace=True)
```

```python
x
```

```python
x
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

**Training** the Model

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(max_depth=32,n_estimators=120,random_state=1)
RF.fit(x_train,y_train)
y_pred = RF.predict(x_test)
```

# Measuring Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
